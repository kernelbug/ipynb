```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
DUMP_PATH = os.path.join('.', 'docs')
URL_MAP_PATH = os.path.join(DUMP_PATH, 'url.map')
LINK_MAP_PATH = os.path.join(DUMP_PATH, 'link.map')
TOKENIZER_PATTERN = re.compile('\W+')
RESIDUAL_PROBABILITY = 0.85
EPS = 1e-6
```

Описатель вершины графа

```python
class Node:
    def __init__(self, url):
        self.url = url
        self.size = None
        self.depth = float('inf')
        self.pr = None
        self.ins = set()
        self.outs = set()
```

## Подсчет всего этого дела

Заведем:
  * _docGraph - соответствие урлов и вершин графа
  * _wordCounter - счетчик слов в коллекции документов

Подразумевается, что скрипты download.py для выкачки Википедии и extractor.py для выделения полезной нагрузки из выкачанных документов уже отработали, поэтому мы можем пройтись по имеющейся коллекции документов, первоначально накидав точек для будущего графа (заполняем _docGraph) и подсчитав частоты слов (_wordCounter) и размеры документов (в Node)

Стоит заметить, что файловая структура коллекции документов такова:
  * n.html - выкачанный скриптом download.py документ
  * n.txt - файл с полезным текстом, полученный скриптом extractor.py из соответствующего n.html
  * url.map - сгенерированное скриптом download.py отображение n.html -> URL
  * link.map - сгенерированное скриптом download.py отображение URL -> out URLs (для построения графа)

```python
_docGraph = {}
_wordCounter = Counter()
with open(URL_MAP_PATH) as urlFile:
    urlMapping = [line.strip().split('\t', 1) for line in urlFile]
for htmlFileName, url in urlMapping:
    htmlFilePath = os.path.join(DUMP_PATH, htmlFileName)
    contentFilePath = os.path.splitext(htmlFilePath)[0] + '.txt'
    node = Node(url)
    node.size = os.path.getsize(htmlFilePath)
    _docGraph[url] = node
    with open(contentFilePath) as contentFile:
        _wordCounter.update(TOKENIZER_PATTERN.split(contentFile.read()))
_docGraph[urlMapping[0][1]].depth = 0
print('Количество токенов: {}'.format(sum(wordCount for wordCount in _wordCounter.values())))
print('Количество уникальных токенов: {}'.format(len(_wordCounter)))
```

Построим граф.

```python
with open(LINK_MAP_PATH) as linkFile:
    for line in linkFile:
        urls = line.strip().split('\t')
        in_ = _docGraph[urls.pop(0)]
        outs = set(_docGraph[url] for url in urls if url in _docGraph)
        in_.outs |= outs
        for out in outs:
            out.ins.add(in_)
_docGraph = list(_docGraph.values())
print('Мощность графа: {}'.format(len(_docGraph)))
```

Вычислим для каждой вершины ее глубину (расстояние данной страницы от главной страницы в кликах), предполагая, что граф связный

```python
nodes = _docGraph[:]
while nodes:
    node = min(nodes, key=lambda node: node.depth)
    depth = node.depth + 1
    for out in node.outs:
        if out.depth > depth:
            out.depth = depth
    nodes.remove(node)
```

Посчитаем PageRank для каждой страницы

```python
residual = []
prGet, prSet = 0, 1
pr = 1 / len(_docGraph)
residualTerm = (1 - RESIDUAL_PROBABILITY) * pr
for node in _docGraph:
    node.pr = [pr, float('inf')]
while True:
    residual.append(max(abs(prNext - prPrev) for prPrev, prNext in (node.pr for node in _docGraph)))
    if residual[-1] < EPS:
        break
    for node in _docGraph:
        node.pr[prSet] = 0
    for node in _docGraph:
        pr = node.pr[prGet] / len(node.outs)
        for out in node.outs:
            out.pr[prSet] += pr
    for node in _docGraph:
        node.pr[prSet] = node.pr[prSet] * RESIDUAL_PROBABILITY + residualTerm
    pr = 1 / sum(node.pr[prSet] for node in _docGraph)
    for node in _docGraph:
        node.pr[prSet] *= pr
    prGet, prSet = prSet, prGet
for node in _docGraph:
    node.pr = node.pr[prGet]
```

```python
plt.plot(range(len(residual)), residual)
plt.grid(True)
plt.xlabel('Iteration')
plt.ylabel('Residual')
plt.title('PageRank convergence')
plt.show()
```

## Построение графиков и выводы

```python
def histogram(data, xLabel, yLabel, title, log=False):
    plt.hist(data, 50, facecolor='green', alpha=0.75)

    if log:
        plt.gca().set_xscale('log')
        plt.gca().set_yscale('log')
    plt.grid(True)
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
    plt.title(title)

    plt.show()
```

Распределение размеров документов в байтах

```python
sizes = [node.size for node in _docGraph]
histogram(sizes, 'Document size (B)', 'Documents', 'Document size distribution')
histogram(sizes, 'Document size (B)', 'Documents', 'Document size distribution (Log)', log=True)
print('Min:', min(sizes))
print('Max:', max(sizes))
```

Распределение входных степеней графа

```python
inDegrees = [len(node.ins) for node in _docGraph]
histogram(inDegrees, 'Inbound degree', 'Documents', 'Inbound degree distribution')
histogram(inDegrees, 'Inbound degree', 'Documents', 'Inbound degree distribution (Log)', log=True)
print('Min:', min(inDegrees))
print('Max:', max(inDegrees))
```

Распределение выходных степеней графа

```python
outDegrees = [len(node.outs) for node in _docGraph]
histogram(outDegrees, 'Outbound degree', 'Documents', 'Outbound degree distribution')
histogram(outDegrees, 'Outbound degree', 'Documents', 'Outbound degree distribution (Log)', log=True)
print('Max:', max(outDegrees))
print('Min:', min(outDegrees))
```

Распределение расстояний страниц от главной страницы

```python
distances = [node.depth for node in _docGraph]
histogram(distances, 'Distance from the main page', 'Clicks', 'Distance from the main page distribution')
print('Min:', min(distances))
print('Max:', max(distances))
```

Распределение частот слов

```python
freqs = [freq for freq in _wordCounter.values()]
histogram(freqs, 'Word frequency', 'Frequency', 'Word frequency distribution', log=False)
histogram(freqs, 'Word frequency', 'Frequency', 'Word frequency distribution (Log)', log=True)
print('Min:', min(freqs))
print('Max:', max(freqs))
```

PageRank Top-20

```python
top = sorted(_docGraph, key=lambda node: node.pr, reverse=True)[:20]
pd.DataFrame.from_dict({'URL': [node.url for node in top], 'PageRank': [node.pr for node in top]})
```
